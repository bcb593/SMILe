```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
#import dmol

soldata = pd.read_csv(
    "https://github.com/whitead/dmol-book/raw/master/data/curated-solubility-dataset.csv"
)
features_start_at = list(soldata.columns).index("MolWt")
#np.random.seed(0)
```

```python
smask = [len(s) <= 96 for s in soldata.SMILES]
```

```python
sum(smask)
```

```python
print(f"Removed {soldata.shape[0] - sum(smask)} long SMILES strings")
filtered_soldata = soldata[smask]
```

```python
filtered_soldata
```

```python
vocab_size=128
```

```python
tokenizer=tf.keras.preprocessing.text.Tokenizer(vocab_size,filters="", char_level=True)
```

```python
tokenizer.fit_on_texts(filtered_soldata.SMILES)
```

```python
seqs = tokenizer.texts_to_sequences(filtered_soldata.SMILES)
```

```python
padded_seqs = tf.keras.preprocessing.sequence.pad_sequences(seqs, padding="post")
```

```python
len(padded_seqs[0])
```

```python
padded_seqs
```

```python
y=filtered_soldata.Solubility
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(padded_seqs,y,test_size=0.2,random_state=101)
X_train,X_val,y_train,y_val=train_test_split(X_train,y_train,test_size=0.4,random_state=42)
```

```python
X_train.shape
```

```python
model = tf.keras.Sequential()

# make embedding and indicate that 0 should be treated as padding mask
model.add(
    tf.keras.layers.Embedding(input_dim=100, output_dim=36, mask_zero=True)
)

# RNN layer
model.add(tf.keras.layers.GRU(32))
# a dense hidden layer
model.add(tf.keras.layers.Dense(32, activation="relu"))
# regression, so no activation
model.add(tf.keras.layers.Dense(1))

model.summary()
```

```python
model.compile(tf.optimizers.Adam(1e-2), loss="mean_squared_error")
result=model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=5, verbose=0)
```
